```python
%matplotlib inline
import matplotlib.pyplot as plt
from statistics import mean
from asyncevo import load
import numpy as np
```

```python
output = load("test.csaga")
my = [mean(step) for step in output['history']]
by = [max(step) for step in output['history']]
```

```python
plt.plot(my, color='red', ls='-')
plt.plot(by, color='red', ls='--')
```

```python
global_sigmas = np.array([[pop['global_sigma'] for pop in step] for step in output['data_history']])
abs_sigmas = np.array([[pop['abs_sigma'] for pop in step] for step in output['data_history']])
abs_paths = np.array([[pop['abs_path'] for pop in step] for step in output['data_history']])
path_norms = np.array([[pop['path_norm'] for pop in step] for step in output['data_history']])
```

```python
plt.plot(np.mean(global_sigmas, axis=0))
```

```python
plt.plot(np.mean(abs_sigmas, axis=0))
```

```python
plt.plot(np.mean(abs_paths, axis=0))
```

```python
plt.plot(np.mean(path_norms, axis=0))
```
