```python
%matplotlib inline
import matplotlib.pyplot as plt
from statistics import mean
from asyncevo import load
import numpy as np
```

```python
output = load("test.csaga")
my = -np.array([mean(step) for step in output['history']])
by = -np.array([max(step) for step in output['history']])
```

```python
plt.semilogy(my, color='red', ls='-')
plt.semilogy(by, color='red', ls='--')
```

```python
local_sigmas = np.array([[pop['local_sigma'] for pop in step] for step in output['data_history']])
global_sigmas = np.array([[pop['global_sigma'] for pop in step] for step in output['data_history']])
abs_sigmas = np.array([[pop['abs_sigma'] for pop in step] for step in output['data_history']])
abs_paths = np.array([[pop['abs_path'] for pop in step] for step in output['data_history']])
path_norms = np.array([[pop['path_norm'] for pop in step] for step in output['data_history']])
```

```python
# local_sigma - the mean abs value of the local component of the step-size update.
plt.plot(np.mean(local_sigmas, axis=1))
```

```python
# global_sigma - the value of the global component of the step-size update.
plt.plot(np.mean(global_sigmas, axis=1))
```

```python
# abs_sigma - the mean of the absolute values of the sigma vector.
plt.plot(np.mean(abs_sigmas, axis=1))
```

```python
# abs_path - the mean of the absolute values of the path vector. This is
# most closely related to the local component of the step-size updated.
plt.plot(np.mean(abs_paths, axis=1))
```

```python
# path_norm - the norm of the path vector.
# This is most closely related to the global component of the step-size update.
plt.plot(np.mean(path_norms, axis=1))
```
