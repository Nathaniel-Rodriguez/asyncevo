```python
%matplotlib inline
import matplotlib.pyplot as plt
from statistics import mean
from asyncevo import load
from asyncevo import Member
```

```python
output = load("test.acsga")
my = [mean(step) for step in output['history']]
by = [max(step) for step in output['history']]
t = output['cooling_schedule'].log
```

```python
fig, ax1 = plt.subplots()
ax1.plot(my, color='tab:blue', ls='-')
ax1.plot(by, color='tab:blue', ls='--')
ax1.set_ylim(-1,0)
ax2 = ax1.twinx()
c = 'tab:red'
ax2.plot(t, color=c)
```
